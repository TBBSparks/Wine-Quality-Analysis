## Import Cleaned CSV's to DF's and Merge

```python
#Dependencies

import pandas as pd
```

### Import Cleaned Salary

```python
# Reference the CSV file

file = "../Resources/CleanedSalary.csv"

#Read the CSV's into a Pandas Dataframe

salary_df = pd.read_csv(file)

#Print header

salary_df.head()
```

```python
#Remove Hyphens from "Occupation_Code" Column so it will match the Automation file

salary_df["Occupation_Code"] = salary_df["Occupation_Code"].replace('-', '', regex=True).astype(int)

#Print header
salary_df.head()
```

### Import Cleaned Automation

```python
# Reference the CSV file

file = "../Resources/CleanedAutomation.csv"

#Read the CSV's into a Pandas Dataframe

automation_df = pd.read_csv(file)

#Print header

automation_df.head()
```

### Merge the Two Dataframes and Export to new CSV

#### Inner Merge

```python
#Inner Join to Only keep files that match

inner_merge_df = pd.merge(salary_df, automation_df, on="Occupation_Code", how = "inner")

#Print head

inner_merge_df.head()
```

```python
#See how many records there are

len(inner_merge_df)
```

```python
#See list of column names

inner_merge_df.columns
```

```python
#Drop columns we don't need

del inner_merge_df['Unnamed: 0_x']
del inner_merge_df['Unnamed: 0_y']
del inner_merge_df['Occupation_Title_y']

inner_merge_df.head()
```

```python
#Rename column


inner_merge_df = inner_merge_df.rename(columns={"Occupation_Title_x":"Occupation_Title"})

inner_merge_df.head()
```

```python
#Export to CSV

inner_merge_df.to_csv("../Resources/CombinedDataInnerMerge.csv", index=True, header=True)
```

#### Outer Merge

```python
#Outer Join to Keep All Files

outer_merge_df = pd.merge(salary_df, automation_df, on="Occupation_Code", how = "outer")

#Print head

outer_merge_df.head()
```

```python
#See how many records there are

len(outer_merge_df)
```

```python
#See list of column names

outer_merge_df.columns
```

```python
#Drop columns we don't need

del outer_merge_df['Unnamed: 0_x']
del outer_merge_df['Unnamed: 0_y']
del outer_merge_df['Occupation_Title_y']

outer_merge_df.head()
```

```python
#Rename column


outer_merge_df = outer_merge_df.rename(columns={"Occupation_Title_x":"Occupation_Title"})

outer_merge_df.head()
```

```python
#Export to CSV

outer_merge_df.to_csv("../Resources/CombinedDataOuterMerge.csv", index=True, header=True)
```
