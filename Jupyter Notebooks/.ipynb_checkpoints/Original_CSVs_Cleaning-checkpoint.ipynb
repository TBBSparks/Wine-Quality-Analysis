## Importing Data and Dependencies

```python
#Import Dependencies

import pandas as pd
```

## Reading in CSV's, Cleaning, and Exporting as a Clean CSV

### Read in Salary CSV and Clean up

```python
# Reference the CSV file

file = "../Resources/Salary.csv"

#Read the CSV's into a Pandas Dataframe

salary_df = pd.read_csv(file)

#Print header

salary_df.head()
```

```python
#Pull out Columns We Want

salary_df = salary_df[["OCC_CODE", "OCC_TITLE", "TOT_EMP", "A_MEAN", "A_PCT10", "A_PCT25", "A_MEDIAN", "A_PCT75", "A_PCT90"]]

#Print Header

salary_df.head()
```

```python
#Rename Columns

salary_df = salary_df.rename(columns={"OCC_CODE":"Occupation_Code",
                                      "OCC_TITLE":"Occupation_Title",
                                      "TOT_EMP":"Total_Employees",
                                      "A_MEAN":"Mean_Annual_Wage",
                                      "A_PCT10":"Annual_10th_Percentile_Wage",
                                      "A_PCT25":"Annual_25th_Percentile_Wage",
                                      "A_MEDIAN":"Annual_Median_Wage",
                                      "A_PCT75":"Annual_75th_Percentile_Wage",
                                      "A_PCT90":"Annual_90th_Percentile_Wage"
                                     })

salary_df.head()
```

```python
#Look for rows with missing data *Looks like we are good

salary_df.count()
```

```python
#Look for duplicates- There are none so no need to drop duplicates

dup_salary_df = salary_df[salary_df.duplicated()]

dup_salary_df
```

```python
#Export cleaned data to new csv

salary_df.to_csv("../Resources/CleanedSalary.csv", index=True, header=True)
```

### Read in Automation CSV and Clean up

```python
#Export cleaned data to new csv

automation_df.to_csv("../Resources/CleanedAutomation.csv", index=True, header=True)
```
